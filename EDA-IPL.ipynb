```python
import os
import pandas as pd
import pandas as pd 
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('whitegrid')
s=os.path.sep
matches_ipl=pd.read_csv("E:"+s+"Astronomy"+s+"Matches IPL 2008-2020.csv")
ipl=pd.read_csv("E:"+s+"Astronomy"+s+"IPL.csv")
ipl.info()
matches_ipl.info()
sns.catplot('extra_runs',data=ipl,hue='extras_type',kind='count')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
bat=ipl["batsman"]
nonbat=ipl["non_striker"]
dism_type=ipl["dismissal_kind"]
match_id=ipl["id"]
count1=0
count2=0
flag1=0
flag2=0
msd_no_mat=0
vk_no_mat=0

for i in range (0,193468,1):
    if(dism_type[i]=='run out'): 
        if(bat[i]=='MS Dhoni' or nonbat[i]=='MS Dhoni'):
            count1=count1+1
    if(bat[i]=='MS Dhoni'or nonbat[i]=='MS Dhoni'):
        flag1=flag1+1
        if(flag1==1):
            msd_no_mat=msd_no_mat+1
    else:     
        flag1=0
    
for i in range (0,193468,1):
    if(dism_type[i]=='run out'): 
        if(bat[i]=='V Kohli' or nonbat[i]=='V Kohli'):
            count2=count2+1  
    if(bat[i]=='V Kohli' or nonbat[i]=='V Kohli'):
        flag2=flag2+1
        if(flag2==1):
            vk_no_mat=vk_no_mat+1
    else:     
        flag2=0   
        
msdpercent=(count1/msd_no_mat)*100     
vkpercent=(count2/vk_no_mat)*100 
print(msdpercent)        
print(vkpercent)   
plotx=['MS Dhoni','V Kohli']
ploty=[msdpercent,vkpercent]
ax.bar(plotx,ploty)
plt.show()        
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
bat_on_str=ipl["batsman"]
bat_on_nonstr=ipl["non_striker"]
batrun=ipl["batsman_runs"]
msd_runs=0
sr_runs=0
for i in range (0,193468,1):
    if(bat_on_str[i]=='MS Dhoni' and bat_on_nonstr[i]=='SK Raina'): 
        msd_runs=msd_runs+batrun[i]
    elif(bat_on_nonstr[i]=='MS Dhoni' and bat_on_str[i]=='SK Raina'):
        sr_runs=sr_runs+batrun[i]
        
plotx=["MS Dhoni","SK Raina"]        
ploty=[msd_runs,sr_runs]
plt.title("Partnership")
ax.bar(plotx,ploty) 
plt.show()
print("Raina and Dhoni contibute equally to their partnership")
```

```python
#Feroz Shah Kotla- Toss Decision
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ven=matches_ipl["venue"]
toss_win=matches_ipl["toss_winner"]
win=matches_ipl["winner"]
toss_dec=matches_ipl["toss_decision"]
chose_bat=0
chose_field=0

for i in range(0,816,1):
    if(ven[i]=='Feroz Shah Kotla'):
        if(win[i]==toss_win[i]):
            if(toss_dec[i]=='bat'):
                chose_bat=chose_bat+1
            else:
                chose_field=chose_field+1
              

            
plotx=['bat','field']
ploty=[chose_bat,chose_field]
plt.title("Feroz Shah Kotla")
ax.bar(plotx,ploty,color='black')            
plt.show()
print("From the plot we understand that teams which won the toss and choose field first have a slightly higher probability of winning the match")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
place=matches_ipl["city"]
toss_win=matches_ipl["toss_winner"]
win=matches_ipl["winner"]
toss_dec=matches_ipl["toss_decision"]
chose_bat=0
chose_field=0

for i in range(0,816,1):
    if(place[i]=='Chennai'):
        if(win[i]==toss_win[i] and toss_win[i]=='Chennai Super Kings'):
            if(toss_dec[i]=='bat'):
                chose_bat=chose_bat+1
            else:
                chose_field=chose_field+1
              

            
plotx=['bat','field']
ploty=[chose_bat,chose_field]
plt.title("Chennai")
ax.bar(plotx,ploty,color='cyan')            
plt.show()
print("From plot we understand that Chennai Super kings have higher probability of winning(at home ground)if they win the toss and choose to bat first")
```
